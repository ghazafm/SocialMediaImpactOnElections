```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop_w = set(STOPWORDS)
```

```python
import pandas as pd 
df = pd.read_csv('D:\BCC\Projek_Intern\SocialMediaSentiment\preprocessing\csv\gabungan.csv')
df.head(10)
```

```python
sw = stopwords.words('indonesian')
df['Tweet'] = df['Tweet'].astype(str)
# keyword = ['ya','yg', 'maju', 'ampa','utk','gak']
# for i in keyword:
#     df['Tweet'] = df['Tweet'].str.replace(i,'')
result = ""
for text in df['Tweet']:
    result += text + " "


print(result)
```

```python
# Membuka file teks baru dalam mode tulis ('w')
with open('all_include.txt', 'w') as file:
    # Tulis string ke dalam file
    file.write(result)
```
