```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/ghazafm/SocialMediaSentiment/main/preprocessing/EDA/id_clean/ganjar.csv')
df.head(10)
```

```python
df_neg = df[df['label']=='Negative']
df_negative = df[df['label']=='Negative']
df_negative = df_negative.groupby('User ID').size().nlargest(10).reset_index(name='Count')
df_negative = pd.DataFrame(df_negative)
df_negative = pd.merge(df_negative, df_neg, on='User ID')
df_negative.to_csv('ganjar_negative.csv')
df_negative['User ID'].value_counts()
df_negative_count = pd.DataFrame(df_negative['User ID'].value_counts())
df_negative_count
```

```python
df_pos = df[df['label']=='Positive']
df_positive = df[df['label']=='Positive']
df_positive = df_positive.groupby('User ID').size().nlargest(10).reset_index(name='Count')
df_positive = pd.DataFrame(df_positive)
df_positive = pd.merge(df_positive, df_pos, on='User ID')
df_positive.to_csv('ganjar_positive.csv')
df_positive_count = pd.DataFrame(df_positive['User ID'].value_counts())
df_positive_count
```
