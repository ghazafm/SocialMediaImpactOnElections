```python
import pandas as pd
import os
```

```python
df_anis = pd.read_csv('https://raw.githubusercontent.com/ghazafm/SocialMediaSentiment/main/preprocessing/EDA/datasets/anis.csv')
df_prabowo = pd.read_csv('https://raw.githubusercontent.com/ghazafm/SocialMediaSentiment/main/preprocessing/EDA/datasets/prabowo.csv')
df_ganjar = pd.read_csv('https://raw.githubusercontent.com/ghazafm/SocialMediaSentiment/main/preprocessing/EDA/datasets/ganjar.csv')
```

```python
keyword = ['inismyname','pilpres','fahnoor','nan','calon','indonesia','survei','survey','pemilu','terkait','gerindra','jokowi','capres','pilih',]

def clean_tweet(content):
    clean = str(content).lower()
    clean = clean.split()
    clean = [word for word in clean if not word in keyword]
    clean = ' '.join(clean)
    return clean
```

```python
def split(data):
    data = str(data)
    data = data.split(" ")
    for i in data:
        i = i.split(" ")
        if i != 'nan':
            list.extend(i)
```

```python
word_anis_n = pd.DataFrame()
word_anis_p = pd.DataFrame()
word_prabowo_n = pd.DataFrame()
word_prabowo_p = pd.DataFrame()
word_ganjar_n = pd.DataFrame()
word_ganjar_p = pd.DataFrame()
```

```python
list = []
word_anis_n = df_anis[df_anis['label'] == 'Negative'] 
word_anis_n = word_anis_n['Clean'].apply(clean_tweet)
word_anis_n = word_anis_n.apply(split)
list = pd.DataFrame(list)
list.dropna(inplace=True)
list.to_csv('word_cloud/anis/negative.csv')

list = []
word_anis_p = df_anis[df_anis['label'] == 'Positive']
word_anis_p = word_anis_p['Clean'].apply(clean_tweet)
word_anis_p = word_anis_p.apply(split)
list = pd.DataFrame(list)
list.dropna(inplace=True)
list.to_csv('word_cloud/anis/positive.csv')
```

```python
list = []
word_prabowo_n = df_prabowo[df_prabowo['label'] == 'Negative']
word_prabowo_n = word_prabowo_n['Clean'].apply(clean_tweet)
word_prabowo_n = word_prabowo_n.apply(split)
list = pd.DataFrame(list)
list.dropna(inplace=True)
list.to_csv('word_cloud/prabowo/negative.csv')

list = []
word_prabowo_p = df_prabowo[df_prabowo['label'] == 'Positive']
word_prabowo_p = word_prabowo_p['Clean'].apply(clean_tweet)
word_prabowo_p = word_prabowo_p.apply(split)
list = pd.DataFrame(list)
list.dropna(inplace=True)
list.to_csv('word_cloud/prabowo/positive.csv')
```
